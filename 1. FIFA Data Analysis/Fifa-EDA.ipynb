## Data Cleaning and Preprocessing

```python

for col in data_df.columns:
    na_val = data_df[col].isna().sum()
    if na_val > 0:
        print('{} --> {}'.format(col, na_val))
```

### Value

The market value of a football player is an estimate of the amount for which a team can sell the player's contract to another team. The longer the contract, the higher the transfer value.

```python
data_df['value'].value_counts() # inspecting the column
```

Here we have to take care of 2 issues

* First the Euro symbol.
* Second coverting the value based on M or K.

```python
def str_to_int(column):
    try:
        value = float(column[1:-1]) # getting the numeric part of the value
        suffix = column[-1] # getting the suffix type
        if suffix == 'M': # if value in M
            value *= 1000000
        elif suffix == 'K': # if value in K
            value *= 1000
    except ValueError: # values other than numeric
        value = 0
    
    return value    
```

```python
data_df['value'] = data_df['value'].apply(str_to_int) # applying the function to convert the data
```

### Wage

```python
# applying the str_to_int() to convert the wage column in the right way

data_df['wage'] = data_df['wage'].apply(str_to_int)
```

### Release Clause

It’s basically a clause inserted into a player’s contract that permits other clubs to attempt to sign that player. Providing of course that they meet the amount in the clause. A release clause is the minimum amount of money you’re willing to accept for that player. 

```python
data_df['release_clause'].value_counts()
```

```python
data_df['release_clause'].fillna(0.0, inplace = True)
```

```python
def str_to_int(column):
    clause = [] # empty list
    for item in data_df[column]: # for every value in the column
        if item != 0: # if item is not 0
            value = float(item[1:-1]) # getting the numeric part of the value
            suffix = item[-1] # getting the suffix type
            if suffix == 'M': # if value in M
                value *= 1000000 # converting to million
                clause.append(value)
            elif suffix == 'K': # if value in K
                value *= 1000 # converting to thousand
                clause.append(value)
        else:
            clause.append(0)
    return clause
```

```python
data_df['release_clause'] = str_to_int('release_clause')
```

```python
data_df['release_clause']
```

```python
# replacing the value 0 to the median of the column

data_df['release_clause'].replace(0, data_df['release_clause'].median(), inplace = True)
```

### Height

```python
data_df['height'].value_counts() # inspecting the height column
```

```python
#function to convert feet+inches to cm

def feet_to_cm(column):
    height = []
    for data in data_df[column]:
        if data != 0: # if data is 0 i.e nan values
            feet, inch = data.split("'")
            feet_to_cm = round(int(feet)*30.48) # mutliplying the feet value with 30.48
            inch_to_cm = round(int(inch)*2.54) # mutliplying the inch value with 2.54
            height.append(feet_to_cm+inch_to_cm) # appeding the total height in cm
        else: # nan value
            height.append(0) # appending 0

    return height
```

```python
data_df['height'] = feet_to_cm('height') # calling the function to perform the transformation
```

```python
data_df['height'].unique() # values after transformation
```

### Postion Columns

```python
position_cols = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm',
 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb']

data_df[position_cols].head()
```

```python
data_df.head()
```

```python
def handling_attribute_values(col):
    replace_col = []
    for item in data_df[col]: # for every value in the column
        if type(item) != float: 
            attribute = item.split('+') # splitting the string on +
            if len(attribute) > 1: # if length of the attribute is greater than 1 i.e a list
                attribute = list(map(int, attribute)) # converting the elements to numeric type
                replace_col.append(sum(attribute)) # appending the sum of the list
            else: # for other values
                replace_col.append(0) # appending 0
        else: # for other types
            replace_col.append(0) # appending 0
    return replace_col
```

```python
for col in position_cols:
    data_df[col] = handling_attribute_values(col)
```

```python
data_df.head(2)
```

### Position

```python
# function to map different position to similar position

def replace_position(df):
    if (df['position'] == 'GK'):
        return 'GK'
    elif ((df['position'] == 'RB') | (df['position'] == 'LB') | (df['position'] == 'CB') | (df['position'] == 'LCB') | (df['position'] == 'RCB') | (df['position'] == 'RWB') | (df['position'] == 'LWB') ):
        return 'DF'
    elif ((df['position'] == 'LDM') | (df['position'] == 'CDM') | (df['position'] == 'RDM')):
        return 'DM'
    elif ((df['position'] == 'LM') | (df['position'] == 'LCM') | (df['position'] == 'CM') | (df['position'] == 'RCM') | (df['position'] == 'RM')):
        return 'MF'
    elif ((df['position'] == 'LAM') | (df['position'] == 'CAM') | (df['position'] == 'RAM') | (df['position'] == 'LW') | (df['position'] == 'RW')):
        return 'AM'
    elif ((df['position'] == 'RS') | (df['position'] == 'ST') | (df['position'] == 'LS') | (df['position'] == 'CF') | (df['position'] == 'LF') | (df['position'] == 'RF')):
        return 'FW'
    else:
        return df.position
```

```python
data_df['position'] = data_df.apply(replace_position, axis = 1)
```
